```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inlineb
%config InlineBackend.figure_format = 'svg'
```

```python
bs_list = [1, 2, 4, 8]
def csv_plot(filename, figsize):
    data = []
    for bs in bs_list:
        data.append(list(pd.read_csv(filename+str(bs)+'_100.csv').iloc[0]))
    fig, ax = plt.subplots(figsize=figsize)
    for idx, row in enumerate(data):
        ax.plot(range(len(row)), row, label='batch_size:' + str(bs_list[idx]))
    plt.legend()
```

```python
csv_plot('resnet152_', (10,5))
```
